```python
import random
from datetime import datetime, timedelta

def generate_random_date(year):
    start_date = datetime(year, 1, 1)  # Start date of the year
    end_date = datetime(year, 12, 31)  # End date of the year
    delta = end_date - start_date
    random_days = random.randint(0, delta.days)  # Generate a random number of days within the year
    random_date = start_date + timedelta(days=random_days)  # Add the random number of days to the start date
    return random_date.strftime('%d-%m-%Y')  # Format the date as dd-mm-yyyy

# Example usage:
# year = 2023  # Change this to your desired year
# random_date = generate_random_date(year)
# print("Random date in {}: {}".format(year, random_date))
# return random_date
```

```python
import random
from datetime import datetime, timedelta

# List of possible login and logout times
login_times = ['8:30 AM', '9:30 AM', '10:30 AM']
logout_times = ['2:30 PM', '3:30 PM', '4:30 PM']

def generate_random_times(base_times):
    random_offset = random.randint(-5, 5)  # Generate a random offset between -5 to 5 minutes
    base_time = datetime.strptime(random.choice(base_times), '%I:%M %p')  # Choose a random base time
    random_time = base_time + timedelta(minutes=random_offset)  # Add the random offset to the base time
    return random_time.strftime('%I:%M %p')  # Format the time as hh:mm AM/PM

def calculate_working_hours(login_time, logout_time):
    login = datetime.strptime(login_time, '%I:%M %p')
    logout = datetime.strptime(logout_time, '%I:%M %p')
    working_hours = logout - login
    return str(working_hours).split(":")[0] + " hours " + str(working_hours).split(":")[1] + " minutes"

# Generate random login and logout times
random_login_time = generate_random_times(login_times)
random_logout_time = generate_random_times(logout_times)

# Calculate working hours
working_hours = calculate_working_hours(random_login_time, random_logout_time)

print("Random login time:", random_login_time)
print("Random logout time:", random_logout_time)
print("Working hours:", working_hours)
```

```python
def get_random_checkin_times():
    random_login_time = generate_random_times(login_times)
    random_logout_time = generate_random_times(logout_times)
    working_hours = calculate_working_hours(random_login_time, random_logout_time)

    return random_login_time, random_logout_time, working_hours
```

```python
checkin_times = []
checkout_times = []
working_hours = []
```

```python
for i in range(checkin_df.shape[0]):
    checkin_time, checkout_time, working_hour = get_random_checkin_times()
    checkin_times.append(checkin_time)
    checkout_times.append(checkout_time)
    working_hours.append(working_hour)
```

```python
import re

def convert_to_hours(working_time):
    # Use regular expression to extract hours and minutes
    match = re.match(r'(\d+) hours (\d+) minutes', working_time)
    if match:
        hours = int(match.group(1))
        minutes = int(match.group(2))
        total_hours = hours + (minutes / 60)  # Convert minutes to fraction of hours
        return int(total_hours)
    else:
        return None  # Return None if the format is not as expected

working_time = "8 hours 04 minutes"
total_hours = convert_to_hours(working_time)

if total_hours is not None:
    print("Total hours:", int(total_hours))
else:
    print("Invalid format for working time.")

```

```python
import pandas as pd

def convert_to_hours(time_str):
    parts = time_str.split()
    hours = int(parts[0])
    minutes = int(parts[2])
    if minutes >= 45:
        hours += 1
    return hours
```
